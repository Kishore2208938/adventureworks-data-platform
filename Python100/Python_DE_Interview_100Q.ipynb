## Section 1: Core Python & Data Structures (Q1–Q20)

### Q1. Given a list of sales transactions, find the top 3 customers by total spend.

```python
from collections import defaultdict

# Question: Find top 3 customers by total spend
transactions = [('C1', 200), ('C2', 150), ('C1', 300),
                ('C3', 500), ('C2', 400), ('C4', 100)]

totals = defaultdict(int)
for cust, amt in transactions:
    totals[cust] += amt

top3 = sorted(totals.items(), key=lambda x: x[1], reverse=True)[:3]
print(top3)
```

**Explanation:** Use `defaultdict` for O(n) aggregation, then sort by value descending and slice top 3. Common pattern for leaderboard/top-N problems.

### Q2. Remove duplicate records from a list of dictionaries based on a specific key ('email').

```python
# Question: Deduplicate list of dicts based on 'email' key
users = [{'id':1,'email':'a@x.com'},
         {'id':2,'email':'b@x.com'},
         {'id':3,'email':'a@x.com'}]

seen = set()
unique = []
for u in users:
    if u['email'] not in seen:
        seen.add(u['email'])
        unique.append(u)
print(unique)
```

**Explanation:** A set tracks seen keys in O(1) while preserving original order. Standard deduplication pattern for staging-layer ETL.

### Q3. Merge two dictionaries by summing values of common keys (e.g., merging daily sales).

```python
from collections import Counter

# Question: Merge two dicts by summing values of common keys
d1 = {'A':100,'B':200}
d2 = {'B':150,'C':300}

merged = dict(Counter(d1) + Counter(d2))
print(merged)
```

**Explanation:** Counter addition merges dicts and sums overlapping keys in one line. Cleaner than manual loops for numeric aggregation.

### Q8. Flatten a nested list of arbitrary depth.

```python
# Question: Flatten a nested list of arbitrary depth

def flatten(lst):
    for item in lst:
        if isinstance(item, list):
            yield from flatten(item)
        else:
            yield item

nested = [1, [2, [3, 4], 5], [6, 7]]
print(list(flatten(nested)))
```

**Explanation:** Recursive generator with `yield from` handles arbitrary nesting without building intermediate lists — memory efficient.

### Q17. Find all pairs in a list that sum to a target value.

```python
# Question: Find all pairs that sum to a target value
nums = [1,2,3,4,5]
target = 6

seen, pairs = set(), []
for n in nums:
    if target - n in seen:
        pairs.append((target - n, n))
    seen.add(n)
print(pairs)
```

**Explanation:** Single-pass O(n) using a set, instead of O(n²) nested loops. Classic two-sum pattern asked in many interviews.

### Q33. Convert a nested dict into a flat dict with dotted keys.

```python
# Question: Flatten a nested dict using dotted keys

def flatten(d, parent=''):
    items = {}
    for k, v in d.items():
        key = f'{parent}.{k}' if parent else k
        if isinstance(v, dict):
            items.update(flatten(v, key))
        else:
            items[key] = v
    return items

d = {'a':1,'b':{'c':2,'d':{'e':3}}}
print(flatten(d))
```

**Explanation:** Standard recursion to flatten JSON payloads into a tabular shape for loading into SQL tables.

### Q38. Find the top-N categories contributing to 80% of revenue (Pareto).

```python
# Question: Find top-N categories contributing to 80% of revenue (Pareto principle)
rev = {'A':500,'B':300,'C':150,'D':50}

sorted_rev = sorted(rev.items(), key=lambda x: -x[1])
total = sum(rev.values())
cumulative, top = 0, []
for name, val in sorted_rev:
    cumulative += val
    top.append(name)
    if cumulative / total >= 0.8:
        break
print(top)
```

**Explanation:** Sort descending, accumulate, then break at the 80% threshold. Classic Pareto analysis for sales/inventory.

### Q39. Normalize a list of numeric values to the 0-1 range (min-max scaling).

```python
# Question: Min-max normalize a list of values to the 0-1 range
values = [10, 20, 30, 40, 50]

lo, hi = min(values), max(values)
normalized = [(v - lo) / (hi - lo) for v in values]
print(normalized)
```

**Explanation:** Min-max is the simplest feature scaling method. Always guard for `hi == lo` to avoid division by zero.

### Q40. Compute a moving average over a window of size k.

```python
# Question: Compute a moving average over a window of size k
lst = [1,2,3,4,5,6]
k = 3

moving = [sum(lst[i:i+k]) / k for i in range(len(lst) - k + 1)]
print(moving)
```

**Explanation:** Sliding window for time-series smoothing. For large data, use a running sum (deque) to reach O(n).

### Q48. Find the 10 most recent files in a directory.

```python
import os, glob

# Question: Find the 10 most recent files in a directory
# Note: Adjust path to a real directory on your system
import tempfile, time

# Create temp files for demo
with tempfile.TemporaryDirectory() as tmpdir:
    for i in range(5):
        path = os.path.join(tmpdir, f'file_{i}.txt')
        with open(path, 'w') as f:
            f.write('data')
        time.sleep(0.01)
    files = glob.glob(os.path.join(tmpdir, '*'))
    recent = sorted(files, key=os.path.getmtime, reverse=True)[:10]
    print([os.path.basename(f) for f in recent])
```

### Q51. Read a .gz compressed CSV file directly.

### Q57. Retry a function up to 3 times on failure (e.g., API call).

```python
import time

# Question: Retry a function up to 3 times on failure with a delay
def retry(func, retries=3, delay=0.1):
    for attempt in range(retries):
        try:
            return func()
        except Exception as e:
            print(f'Attempt {attempt+1} failed: {e}')
            if attempt == retries - 1:
                raise
            time.sleep(delay)

# Demo: function that fails twice then succeeds
call_count = [0]
def flaky():
    call_count[0] += 1
    if call_count[0] < 3:
        raise ConnectionError("timeout")
    return "success"

result = retry(flaky)
print(result)
```

**Explanation:** Re-raise on the final attempt so the caller still sees failures. In production use `tenacity` for backoff and jitter.

### Q61. Implement a decorator to log function execution time.

```python
import time
from functools import wraps

# Question: Implement a decorator that logs the execution time of a function
def timing(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        print(f'{func.__name__}: {time.time()-start:.3f}s')
        return result
    return wrapper

@timing
def etl_job():
    time.sleep(0.1)

etl_job()
```

**Explanation:** `@wraps` preserves metadata of the wrapped function. Common pattern for profiling ETL jobs without polluting logic.

### Q83. ETL: Split a large CSV into smaller chunks of N rows each.

```python
import csv, io

# Question: ETL — Split a large CSV into chunks of N rows each

def split_csv(csv_str, n=2):
    reader = csv.reader(io.StringIO(csv_str))
    header = next(reader)
    chunk, idx = [], 1
    for row in reader:
        chunk.append(row)
        if len(chunk) == n:
            buf = io.StringIO()
            w = csv.writer(buf)
            w.writerow(header)
            w.writerows(chunk)
            print(f'=== chunk_{idx}.csv ===')
            print(buf.getvalue())
            chunk, idx = [], idx + 1
    if chunk:
        buf = io.StringIO()
        w = csv.writer(buf)
        w.writerow(header)
        w.writerows(chunk)
        print(f'=== chunk_{idx}.csv ===')
        print(buf.getvalue())

csv_data = "id,name\n1,A\n2,B\n3,C\n4,D\n5,E"
split_csv(csv_data, n=2)
```

**Explanation:** Chunking enables parallel loads and fits oversized files into downstream row-limit APIs.

**Explanation:** Simple threshold monitoring is the foundation of real-time alerting. Extend with time windows for rate-based alerts.

### Q100. ETL: Build an end-to-end mini pipeline — extract, transform, load.

```python
import csv, io

# Question: ETL — Build a complete mini ETL pipeline with separate extract, transform, load steps

def extract(csv_str):
    return list(csv.DictReader(io.StringIO(csv_str)))

def transform(rows):
    for r in rows:
        r['amount'] = float(r['amount'])
    return [r for r in rows if r['amount'] > 0]

def load(rows):
    buf = io.StringIO()
    w = csv.DictWriter(buf, fieldnames=rows[0].keys())
    w.writeheader()
    w.writerows(rows)
    return buf.getvalue()

# Run the pipeline
raw_csv = "order_id,amount\n1,500\n2,-10\n3,300\n4,0"
result = load(transform(extract(raw_csv)))
print(result)
```

**Explanation:** Separating E/T/L into pure functions makes each step unit-testable and reusable. Core design pattern for any Python ETL.
